```python
import numpy as np

a = np.array([[1, 6], [2, 8], [3, 11], [3, 10], [1, 7]], dtype=np.float64)
mean_a = np.array(a.mean(axis=0))
a_centered = a - mean_a
a_centered_sp = a_centered.T[0] @ a_centered.T[1]
cov_a = a_centered_sp / (a_centered.shape[0]-1)
cov_a
```
