```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from pylab import rcParams
plt.style.use('fivethirtyeight')
```

```python
card_df = pd.read_csv('creditcard.csv', sep=',')
```

```python
card_df.head(10)
```

```python
card_df.info()
```

```python
card_df['Class'].value_counts()
```

```python
card_df['Class'].value_counts().to_frame()
```

```python
classes = card_df['Class'].value_counts()
classes.plot.bar()
rcParams['figure.figsize'] = '8.75, 4.5'
```

```python
classes = card_df['Class'].value_counts()
classes.plot.bar(logy=True)
rcParams['figure.figsize'] = '9.25, 4.5'
```

```python
title_dict = {'fontsize': 14, 'fontweight': 'bold', 'color': '#404040', 'family':'Ubuntu'}
label_dict = {'fontsize': 12, 'fontweight': 'normal', 'color': '#808080', 'family':'Noto mono'}

plt.hist(card_df[card_df['Class'] == 1]['V1'], bins=20, alpha=0.5, color='#df0000', density=True)
plt.hist(card_df[card_df['Class'] == 0]['V1'], bins=20, alpha=0.5, color='#909090', density=True)
plt.xlim(-32, 3)
rcParams['figure.figsize'] = '9.25, 4.5'
plt.xlabel('V1', fontdict=label_dict)
plt.legend(labels = ['Class 0', 'Class 1'], loc='best')
```
