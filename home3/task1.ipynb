### задание 1

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston.data
```

```python
data.shape
```

```python
data
```

```python
target = boston.target
```

```python
feature_names = boston.feature_names
feature_names
```

#### ниже идет описание датасета

```python
for line in boston.DESCR.split('\n'):
    print(line)
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
X.shape
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=['price'])
y.head(10)
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten()}, columns=['y_test', 'y_pred'])
check_test.head(10)
```

#### R2

```python
from sklearn.metrics import r2_score
```

```python
print(f'R2: {r2_score(y_test, y_pred)}')
```

### задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RndF = RandomForestRegressor(n_estimators=1000, max_depth=14, random_state=42)
```

```python
RndF.fit(X_train, y_train.values[:,0])
```

```python
y_pred = RndF.predict(X_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten()}, columns=['y_test', 'y_pred'])
check_test.head(10)
```

```python
print(r2_score(y_test, y_pred))
```

#### RandomForestRegressor отработал лучше, его значение R2 = 0.8748718718512973 ближе к единице. R2 у LinearRegression  0.711226005748496
